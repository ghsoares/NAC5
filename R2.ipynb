### Ângulo entre os círculos

```python
import math
from helper_functions import draw_text

# Calcula diferença entre as posições dos círculos
diff_x = (red_center_x - blue_center_x)
diff_y = (red_center_y - blue_center_y)

# Centro entre os círculos
center_x = int((red_center_x + blue_center_x) * .5)
center_y = int((red_center_y + blue_center_y) * .5)

# Calcula o ângulo em radianos usando arcotangente
ang_rad = math.atan2(diff_y, diff_x)

# Converte de radianos para graus 
ang_deg = ang_rad * (180.0 / math.pi)

# Desenha a reta de origem
cv2.line(
	cont_img, (0, center_y),(img_size[1], center_y),
	(0, 0, 0), 8
)

# Desenha um arco do ângulo
cv2.ellipse(
	cont_img, (center_x, center_y), (150, 150), 0, 0, ang_deg,
	(150, 64, 42), 10
)

# Desenha o texto do ângulo
draw_text(
	cont_img, (center_x + 150, center_y - 16),
	cv2.FONT_HERSHEY_SIMPLEX, str(int(ang_deg)) + "d", (0, 0, 0)
)

# Mostra o resultado
plt.imshow(cont_img)
plt.show()
```
